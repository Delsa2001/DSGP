```python
import os
import re

def rename_files(directory):
    for root, _, files in os.walk(directory):
        for idx, filename in enumerate(files):
            # Remove non-ASCII characters and add a unique index to each file name
            new_filename = re.sub(r'[^\x00-\x7F]+', '_', filename)
            new_filename = f"{idx}_{new_filename}"  # Add an index to ensure uniqueness
            
            old_path = os.path.join(root, filename)
            new_path = os.path.join(root, new_filename)
            
            # Rename the file if the name has changed
            if old_path != new_path:
                try:
                    os.rename(old_path, new_path)
                    print(f"Renamed: {old_path} -> {new_path}")
                except FileExistsError:
                    print(f"File already exists, skipping: {new_path}")

# Use your data_path here
rename_files('E:/IIT Year 2/Sem 1/DSGP/Images')

```
